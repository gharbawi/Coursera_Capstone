<h1>Project name:</h1>
<h3><i>Your Road</i></h3>

<h2>Abstract</h2>

<p>This project is describing the implementation of machine learning in the way of how the tourists could choose thier trip. Most of web sites might gives you a recommendation includes the price of hotels, shortest road and other useful things. However, it is still missing giving you a most convenient recommendation which fit your style,requirments, mode, needs or even gives you somthing difference regardless the time or the cost.</p>

<p> The project will use unsupervised machine learning method, Clustring, using K means as a way to cluster the venues of the predetermind latitude or langtitude of the proposed trip. then evaluate if it fit tourist profile or not. a map with the clusters will help the toursit to choose other better roads passing in areas fit his requirments.

<p> Data is available by Foursquare location data provider. By exploring the venues around the located latitude and longtitude. the category for each venues will be provided then form a data frame including these venues and categories. Data exploratory and analysis are a long process to build a data structure fit to the model</p>

<p> In conclusion, the tourist or the new commers will get a simple tool which will help the tourist to choose different roads for his target through a recommendations of machinelearning clustering application <b><i> " Your Road"</i></b>
